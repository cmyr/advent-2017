```python
def location_for_idx(idx):
    h = math.ceil(math.sqrt(idx))
    if h % 2 == 0:
        h = h + 1
    
    # h = the length of an arm at this width
    
    # w = the straight-line distance from the center to this arm of the spiral
    w = (h-1) / 2
    
    
    # d = the distance from the center to the end of an arm
    d = w*2

#     print('w: %s, h: %s, d: %s' % (w, h, d))

#     w_prev = h - 2
#     print('w prev: ', w_prev)
    
    # this idx, offset from the start of this layer
    rel_idx = idx - (h-2)**2
#     print('rel_idx: ', rel_idx)
#     print(idx % w**2)
    x, y = 0, 0
    for i in range(1, 5):
        if rel_idx <= i * d:
            assert math.ceil(rel_idx / d) == i, "%s, %s" % (i, rel_idx / d)
            if i == 1:
                x = w
                y = rel_idx - w
                break
            elif i == 2:
                y = w
                x = w - (rel_idx - d)
                break
            elif i == 3:
                x = - w
                y = w - (rel_idx - d*2)
#                 print('huh', w, x, y)
                break
            elif i == 4:
                y = - w
                x = (rel_idx - d*3) - w
                break
            else:
                assert false, "you shouldn't be here"
#     print('i: ', i)
    return (x, y)

x = location_for_idx(6); assert x == (-1, 0), x
x = location_for_idx(17); assert x == (-2, 2), x
x = location_for_idx(21); assert x == (-2, -2), x
x = location_for_idx(49); assert x == (3, -3), x
x = location_for_idx(31); assert x == (3, 3), x
```

```python
def index_for_location(x, y):
    w = max(abs(x), abs(y))
    h = w*2 + 1
    
    max_for_width = h**2
    min_for_width = (h-2)**2
#     print('w: ', w, 'max: ', max_for_width)
    adjustment = 0
    if x == w and y == - w:
        return int(max_for_width)
    elif x == w:
        return int(min_for_width + w + y)
    elif y == w:
        return int(min_for_width + w*3 + x * -1)
    elif x == -w:
        return int(min_for_width + w*5 + y * -1)
    else:
        return int(min_for_width + w*7 + x)
    

print(index_for_location(-1, 0))
print(index_for_location(-2, 2))
print(index_for_location(2, -1))
print(index_for_location(3, -3))
print(index_for_location(*location_for_idx(25)))
print(index_for_location(*location_for_idx(24)))
print(index_for_location(*location_for_idx(23)))
    
```
